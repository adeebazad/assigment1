Q5-Is Python code compiled or Interpreted?

Ans5- Python is an interpreted language, not a compiled language. This means that the source code is executed line by line, at runtime, by an interpreter, rather than being compiled into machine code ahead of time, which is executed directly by the computer's hardware.

Q15-How to Install Python on Windows and set path variables?

Ans15- To set the path variables, we will need to update the system environment variables.
To do this,these steps:
Press the Windows key and the X key at the same time.
Select "System."
Click on "Advanced system settings."
Click on the "Environment Variables" button.
Under "System Variables," scroll down and find the "Path" variable.
Click on the "Edit" button.
Click on the "New" button and add the path to the Python installation, for example: C:\Python38\.
