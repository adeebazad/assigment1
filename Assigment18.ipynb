Q1. What is Flask Framework? What are the advantages of Flask Framework?


Ans-

Flask is a micro web framework written in Python that allows developers to build web applications quickly and with minimal boilerplate code. It is designed to be simple, lightweight, and easy to learn, making it a popular choice for small to medium-sized web applications and APIs.

Advantages of using Flask framework include:

Lightweight and minimalistic: Flask is designed to be simple and lightweight, with a small core and modular design that allows developers to add only the features they need. This makes it easy to learn and use, and reduces the amount of code needed to get a project up and running.

Flexible and customizable: Flask is highly customizable and can be easily extended with third-party libraries to add additional functionality. This allows developers to tailor their applications to their specific needs and requirements.

Easy to learn and use: Flask has a simple and intuitive API that makes it easy to learn and use, even for beginners. The documentation is comprehensive and easy to follow, and the community is active and supportive.

Integrated testing support: Flask comes with built-in support for testing, making it easy to write and run unit tests for your application. This helps ensure that your application is functioning correctly and reduces the risk of bugs and errors.

Q3. What is App routing in Flask? Why do we use app routes?


Ans-

app routing refers to the process of mapping URLs to view functions. When a user makes a request to a URL, Flask uses the app routing system to determine which view function should be called to handle the request.

We use app routes in Flask to define the behavior of our web application. App routes allow us to define which URLs our application should respond to and what should happen when a user visits each URL. We can use app routes to display HTML templates, return JSON data, redirect to other pages, and perform other actions based on the user's request.

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

Ans-

In Flask, the url_for() function is used for URL building. It generates a URL for a given endpoint with the optional method and arguments.


```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'This is the homepage.'

@app.route('/user/<username>')
def show_user_profile(username):
    return 'User %s' % username

if __name__ == '__main__':
    with app.test_request_context():
        print(url_for('index'))
        print(url_for('show_user_profile', username='pwskill'))
```
