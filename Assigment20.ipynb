Q1. 
Ans-GET and POST methods are HTTP methods used to send data from a client to a server. In Flask, GET method is used to request data from a specified resource while POST method is used to submit an entity to a specified resource. The request object in Flask contains all the data sent from the client to the server and this data can be recovered using the GET/POST methods1.


Q2. 
Ans-The request object in Flask is used to access incoming request data. It is an instance of the Request class and contains information about incoming requests such as headers, method, URL etc2.

Ans -Templates in Flask are files that contain static data as well as placeholders for dynamic data. A template is rendered with specific data to produce a final document. Flask uses the Jinja template library to render templates. The render_template() function both selects the template file to be used and passes any values or variables it needs4.
